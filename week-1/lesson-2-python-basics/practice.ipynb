## 1. Variables

````python
# A variable is a box with a name. Text needs quotes; numbers do not.
name = "Ali"
age = 25
price = 19.99
print(name, age, price)
````

````python
# You can change what is inside a box.
age = 25
age = age + 1     # add 1
print(age)        # 26
````

**Try it:** change `name` to your own name and run again.

## 2. Data types

````python
# type() tells you the kind of value.
print(type("Tashkent"))   # str  (text)
print(type(42))           # int  (whole number)
print(type(19.99))        # float (decimal)
print(type(True))         # bool  (True/False)
````

## 3. Text is not a number

````python
age_text = "30"
print(int(age_text) + 5)   # 35 -> first change "30" into 30, then add
print(str(30) + " years")  # "30 years" -> change number to text to join
````

**Try it:** what happens if you run `"30" + 5` ? (It gives an error — that is expected.)

## 4. Math and comparison

````python
print(7 + 2)    # 9
print(7 / 2)    # 3.5  (normal divide)
print(7 // 2)   # 3    (drop the decimal)
print(7 % 2)    # 1    (the remainder)
````

````python
print(5 == 5)   # True  (equal?)
print(5 != 3)   # True  (not equal?)
print(10 > 20)  # False
````

## 5. f-strings

````python
name = "Ali"
age = 25
print(f"{name} is {age} years old")    # Ali is 25 years old

price = 19.876
print(f"Price: {price:.2f}")           # Price: 19.88  (2 decimals)
````

## 6. if / elif / else

````python
score = 75
if score >= 90:
    print("excellent")
elif score >= 60:
    print("pass")
else:
    print("fail")
````

````python
# join questions with and / or / not
tenure = 4
charges = 80
if tenure < 6 and charges > 50:   # both True
    print("at risk")
else:
    print("safe")
````

**Try it:** change `score` to 95, then to 40, and run again.

## 7. for loop

````python
total = 0
for n in [5, 10, 15, 20]:
    total = total + n
print(total)        # 50
````

````python
# range(5) gives 0,1,2,3,4
for i in range(5):
    print(i)
````

````python
# break stops the loop, continue skips one item
for n in [1, 2, 3, 4, 5]:
    if n == 4:
        break          # stop when we reach 4
    print(n)           # prints 1 2 3
````

## 8. list

````python
prices = [10, 20, 30, 40, 50]
print(prices[0])      # 10  -> first item (index 0)
print(prices[-1])     # 50  -> last item
print(prices[1:3])    # [20, 30] -> index 1 and 2 (stop before 3)
print(len(prices))    # 5   -> how many items
````

````python
prices.append(60)     # add 60 to the end
print(prices)         # [10, 20, 30, 40, 50, 60]
````

## 9. dictionary

````python
customer = {"name": "Ali", "city": "Tashkent", "age": 30}
print(customer["name"])               # Ali
print(customer.get("plan", "basic"))  # basic (no 'plan' key -> default)
print(customer.keys())                # all keys
print(customer.values())              # all values
````

````python
# loop over a dictionary
for key in customer:
    print(key, "->", customer[key])
````

## 10. function

````python
def price_with_tax(x, tax=10):    # tax has a default of 10
    return x + x * tax / 100
print(price_with_tax(200))       # 220
print(price_with_tax(200, 5))    # 210
````

````python
# a function with no return gives back None
def say_hi(name):
    print(f"Hi {name}")
result = say_hi("Zara")     # prints Hi Zara
print(result)               # None  (because there is no return)
````
