## numpy

```python
import numpy as np
a = np.array([10, 20, 30, 40, 50])
print(a + 10)       # add 10 to each -> [20 30 40 50 60]
print(a * 2)        # multiply each by 2
print(a.mean())     # the average
```

```python
# boolean mask: keep only numbers bigger than 25
print(a[a > 25])    # [30 40 50]
```

## pandas - load the real data

```python
import pandas as pd
url = "https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv"
df = pd.read_csv(url)
print("rows, columns:", df.shape)
df.head()
```

```python
# column names and types
df.info()
```

## pick a column, filter rows

```python
print(df["Contract"].value_counts())   # how many of each contract type
```

```python
# keep only customers with tenure under 6 months
new = df[df["tenure"] < 6]
print("new customers (tenure < 6):", len(new))
```

## missing values and groupby

```python
print(df.isna().sum().sort_values(ascending=False).head(5))
```

```python
# average monthly charge for each contract type
print(df.groupby("Contract")["MonthlyCharges"].mean().round(2))
```
