```python
import pandas as pd
url = "https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv"
df = pd.read_csv(url)
print("shape:", df.shape)
```

## 1. TotalCharges looks numeric but is text

```python
print("type before:", df["TotalCharges"].dtype)   # object (text)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
print("type after :", df["TotalCharges"].dtype)    # float64
print("missing now:", df["TotalCharges"].isna().sum())
```

## 2. Find missing values

```python
print(df.isna().sum().sort_values(ascending=False).head(5))
```

## 3. Fill missing numbers with the median

```python
m = df["TotalCharges"].median()
df["TotalCharges"] = df["TotalCharges"].fillna(m)
print("missing after fill:", df["TotalCharges"].isna().sum())   # 0
```

## 4. Duplicates and unrealistic values

```python
print("duplicate rows:", df.duplicated().sum())
print(df[["tenure", "MonthlyCharges", "TotalCharges"]].describe())
```
