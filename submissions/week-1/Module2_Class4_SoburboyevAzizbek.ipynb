````python
# === SETUP — run this first ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings('ignore')

url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv'
df = pd.read_csv(url)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)
print('Loaded:', df.shape)
````

### Cell 1 — see the first rows

````python
df.head() #boshidagi 5 ta rowni ko'rsatadi
````

### Cell 2 — dimensions of the data

````python
df.shape # ko'rsatadi qator va ustunlarni
````

### Cell 3 — column names and types

````python
df.dtypes # ustunlarni turini ko'rsatadi
````

### Cell 4 — count customers who churned vs stayed

````python
df['Churn'].value_counts() # churnda nechta yes va no borligini aytadi
````

### Cell 5 — only customers with tenure > 50 months

````python
long_term = df[df['tenure'] > 50]
print('Long-term customers:', len(long_term))
long_term.head()

# qiymati 50 tadan katta tenure sonni print qiladi va daslabki 5 tasini jadvalini ko'rsatadi
````

### Cell 6 — top 5 customers by TotalCharges

````python
top5 = df.sort_values('TotalCharges', ascending=False).head(5)
top5[['customerID', 'tenure', 'MonthlyCharges', 'TotalCharges']]

# jadval yaratadi custoomer id, tenure, monthlycharges, totalcharges ustunlarini. dastlabki 5 ta qatorni ko'rsatadi va totalcharges ustunini kamayish tartibinda saralangan holda
````

### Cell 7 — average MonthlyCharges per contract type

````python
df.groupby('Contract')['MonthlyCharges'].mean().round(2)

# contractdagi turlari bo'yicha sonlarni o'rtacha qiymatini ko'rsatadi, round bu yaxlitlash
````

### Cell 8 — average MonthlyCharges AND tenure per Churn group

````python
df.groupby('Churn').agg({'MonthlyCharges': 'mean', 'tenure': 'mean'}).round(2)
# Churndagi yes va no larnining oylik o'rtacha qiymatini va tenureni ham o'rtacha qiymatini exselga qo'shadi
````

### Cell 9 — count of senior citizens by Churn

````python
pd.crosstab(df['SeniorCitizen'], df['Churn'])
#crosstab sanaydi, yani yoshi kattalardan nechtasi yes va no, yoshi kichiklardan nechtasi yes va no larni exsel shaklida ko'rsatadi 
````
